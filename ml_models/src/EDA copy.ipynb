Las principales razones operacionales y académicas detrás de un reingreso secuencial (abrir un nuevo `COD_ALUMNO` en un período posterior) suelen ser:

- **Cambio de Carrera:** La persona estudió la Carrera A, la abandonó o se retiró formalmente, y luego regresó años más tarde para estudiar la Carrera B (o la misma carrera, pero en un plan de estudios diferente), lo que obligó al sistema a generar un nuevo expediente (`COD_ALUMNO`).

- **Reingreso Tardo (Reincorporación):** El estudiante interrumpió sus estudios por un período prolongado y, al reincorporarse, la normativa institucional exigió la apertura de un nuevo expediente en lugar de reactivar el anterior.

- **Postgrado/Segunda Especialidad:** Aunque menos común, algunos sistemas de información académica asignan un `COD_ALUMNO` distinto para estudios de postgrado o segunda especialidad, aunque la persona ya tenga un `COD_PERSONA` activo.

debido a la presencia predominante de nulos en periodos de ingreso particulares, a pesar que el porcentaje de nulos sea 2.81% se realizara la imputacion.

El metodo de imputacion por fechas es complicado, por ello nos apoyaremos en el periodo de ingreso, calcularemos la edad promedio con la que un estudiante promedio ingresa a la universidad, y luego restaremos dicha edad al periodo de ingreso, obteniendo una imputacion de la fecha de nacimiento 

**Lógica**
- Agrupar por persona (COD_PERSONA), ya que puede tener varios COD_ALUMNO.
- Obtener la primera ocurrencia de ingreso (el periodo más antiguo, porque representa su ingreso inicial).
- Mantener la fecha de nacimiento si existe, y si no, se imputará luego.
- Calcular la edad promedio de ingreso entre quienes sí tienen fecha válida.
- Imputar las fechas faltantes restando esa edad promedio (en años) al periodo de ingreso.

````python
import pandas as pd
import numpy as np

# 1️⃣ Convertir la fecha de nacimiento a string, eliminando decimales y nulos
df['FECHA_NACIMIENTO'] = (
    df['FECHA_NACIMIENTO']
    .dropna()
    .astype(int)
    .astype(str)
)

# 2️⃣ Convertir al tipo datetime con formato AAAAMMDD
df['FECHA_NACIMIENTO'] = pd.to_datetime(df['FECHA_NACIMIENTO'], format='%Y%m%d', errors='coerce')

# 1️⃣ Subconjunto relevante
df_sub = df[['COD_PERSONA', 'FECHA_NACIMIENTO', 'PER_INGRESO']].copy()
print(df_sub.head())

# 2️⃣ Convertir fecha de nacimiento
df_sub['FECHA_NACIMIENTO'] = pd.to_datetime(df_sub['FECHA_NACIMIENTO'], errors='coerce')
print(df_sub.head())
# 3️⃣ Convertir PER_INGRESO a fecha estimada según semestre
def periodo_a_fecha(periodo):
    try:
        anio, sem = periodo.split('-')
        anio = int(anio)
        sem = int(sem)
        if sem == 1:
            return pd.Timestamp(f"{anio}-03-01")  # primer semestre → marzo
        elif sem == 2:
            return pd.Timestamp(f"{anio}-08-01")  # segundo semestre → agosto
    except:
        return pd.NaT

df_sub['PER_INGRESO_FECHA'] = df_sub['PER_INGRESO'].apply(periodo_a_fecha)

# 4️⃣ Tomar el primer periodo de ingreso por persona
df_min = df_sub.groupby('COD_PERSONA', as_index=False).agg({
    'FECHA_NACIMIENTO': 'first',
    'PER_INGRESO_FECHA': 'min'
})

df_min.info()
df_min.head()

# 5️⃣ Calcular edad en años completos
df_valid = df_min.dropna(subset=['FECHA_NACIMIENTO']).copy()
df_valid['EDAD_INGRESO'] = df_valid.apply(
    lambda x: x['PER_INGRESO_FECHA'].year - x['FECHA_NACIMIENTO'].year -
    ((x['PER_INGRESO_FECHA'].month, x['PER_INGRESO_FECHA'].day) <
     (x['FECHA_NACIMIENTO'].month, x['FECHA_NACIMIENTO'].day)),
    axis=1
)

# 6️⃣ Calcular edad promedio
edad_promedio = round(df_valid['EDAD_INGRESO'].mean())

print(f"Edad promedio de ingreso: {edad_promedio} años")
# df_valid cuando la edad es menor a 15 años
print(df_valid['EDAD_INGRESO'].describe())
print(df_valid[df_valid['EDAD_INGRESO'] < 15])
print (df_valid['EDAD_INGRESO'].describe())
# barplot de edad ingreso
plt.figure(figsize=(10, 6))
sns.histplot(df_valid['EDAD_INGRESO'], bins=30, kde=True, color='skyblue')
plt.title('Distribución de Edad de Ingreso', fontsize=16)
plt.xlabel('Edad de Ingreso (años)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.axvline(df_valid['EDAD_INGRESO'].mean(), color='red', linestyle='--', label='Media')
plt.axvline(df_valid['EDAD_INGRESO'].median(), color='green', linestyle='-', label='Mediana')
plt.legend()
plt.grid(axis='y', linestyle='dotted', alpha=0.6)
plt.show()
print(df_valid['EDAD_INGRESO'].value_counts().sort_index())

````

````python
# 7️⃣ Imputar fechas faltantes o edades irreales (<10 años)
df_min['FECHA_NACIMIENTO_IMPUTADA'] = df_min['FECHA_NACIMIENTO']

# Personas con edad < 10
cod_menores_10 = df_valid.loc[df_valid['EDAD_INGRESO'] < 10, 'COD_PERSONA']

# Máscara combinada: nulos o menores de 10
mask = df_min['FECHA_NACIMIENTO'].isna() | df_min['COD_PERSONA'].isin(cod_menores_10)

# Imputar usando la edad promedio
df_min.loc[mask, 'FECHA_NACIMIENTO_IMPUTADA'] = df_min.loc[mask, 'PER_INGRESO_FECHA'] - pd.DateOffset(years=edad_promedio)

# 🔚 Resultado final
df_final = df_min[['COD_PERSONA', 'PER_INGRESO_FECHA', 'FECHA_NACIMIENTO_IMPUTADA']]
````

### PROCEDENCIA Y RESIDENCIA

````python
# valores nulos de PROCEDENCIA
print('Cantidad de valores nulos en PROCEDENCIA:', df['PROCEDENCIA'].isnull().sum())
# valores unicos de PROCEDENCIA
print('Cantidad de valores unicos en PROCEDENCIA:', df['PROCEDENCIA'].nunique())

# valores nulos de RESIDENCIA
print('Cantidad de valores nulos en RESIDENCIA:', df['RESIDENCIA'].isnull().sum())
# valores unicos de RESIDENCIA
print('Cantidad de valores unicos en RESIDENCIA:', df['RESIDENCIA'].nunique())
````

Se identificó que no existen datos faltantes en el conjunto. Por otro lado, la variable `PROCEDENCIA` presenta **158 valores únicos**, lo cual la hace poco conveniente para ser usada directamente en un modelo, pues su alta cardinalidad podría afectar su desempeño. Además, considerando que a nivel nacional hay más de **1,800 distritos**, esta cifra podría seguir creciendo con nuevos datos, lo que la haría aún menos práctica.

Por ello, se propone **extraer nuevas características** a partir de la dirección de procedencia y residencia, haciendo uso de fuentes estadísticas oficiales del Perú:

- **Porcentaje de pobreza total por distrito**: se obtendrá del Mapa de Pobreza del INEI. [(INEI – Mapa de Pobreza)](https://www.gob.pe/institucion/inei/colecciones/6608-mapa-de-pobreza)  
- **Índice de Desarrollo Humano (IDH) distrital**: se obtendrá de informes del PNUD / INEI que calculan el IDH para distritos peruanos. (por ejemplo, “Mapas Índice de desarrollo humano a nivel distrital” – MINAM / PNUD) [(Perú: Índice de Desarrollo Humano Distrital 2005.)](https://sinia.minam.gob.pe/documentos/peru-indice-desarrollo-humano-distrital-2005)

Con estas nuevas variables se podrá capturar la dimensión socioeconómica del lugar de procedencia, reduciendo la cardinalidad del feature `PROCEDENCIA` y `RESIDENCIA` y agregando información útil al modelo.

Para ello, antes dividiremos `PROCEDENCIA` en `DEPARTAMENTO_PRO`,`PROVINCIA_PRO`, `DISTRITO_PRO` y `RESIDENCIA` en `DEPARTAMENTO_RES`,`PROVINCIA_RES`, `DISTRITO_RES`.



````python
# División de PROCEDENCIA
df[['DEPARTAMENTO_PRO', 'PROVINCIA_PRO', 'DISTRITO_PRO']] = df['PROCEDENCIA'].str.split('/', expand=True)

# División de RESIDENCIA
df[['DEPARTAMENTO_RES', 'PROVINCIA_RES', 'DISTRITO_RES']] = df['RESIDENCIA'].str.split('/', expand=True)

# Función para limpiar texto: minúsculas + sin acentos + sin espacios extra
def limpiar_texto(col):
    return col.str.upper().apply(lambda x: unidecode(x.strip()) if isinstance(x, str) else x)

# Aplicar limpieza a todas las nuevas columnas
cols_a_limpiar = [
    'DEPARTAMENTO_PRO', 'PROVINCIA_PRO', 'DISTRITO_PRO',
    'DEPARTAMENTO_RES', 'PROVINCIA_RES', 'DISTRITO_RES'
]

df[cols_a_limpiar] = df[cols_a_limpiar].apply(limpiar_texto)
````

````python
# valores unicos de departamento en DEPARTAMENTO_PRO y DEPARTAMENTO_RES
print('Valores unicos en DEPARTAMENTO_PRO:', df['DEPARTAMENTO_PRO'].unique())
print('Valores unicos en DEPARTAMENTO_RES:', df['DEPARTAMENTO_RES'].unique())
````

Al evaluar los valores unicos de `DEPARTAMENTO` nos damos cuenta que si existia valores unicos, solo que estaban bajo el string `//`
por lo tanto se evaluara la magnitud de los nulos, y su posible manejo.

````python
# transformar '' en NaN
df['DEPARTAMENTO_PRO'] = df['DEPARTAMENTO_PRO'].replace('', np.nan)
df['PROVINCIA_PRO'] = df['PROVINCIA_PRO'].replace('', np.nan)
df['DISTRITO_PRO'] = df['DISTRITO_PRO'].replace('', np.nan)
df['DEPARTAMENTO_RES'] = df['DEPARTAMENTO_RES'].replace('', np.nan)
df['PROVINCIA_RES'] = df['PROVINCIA_RES'].replace('', np.nan)
df['DISTRITO_RES'] = df['DISTRITO_RES'].replace('', np.nan)

# cantidad de nulos en DEPARTAMENTO
print('Cantidad de valores nulos en DEPARTAMENTO_PRO:', df['DEPARTAMENTO_PRO'].isnull().sum())
print('Cantidad de valores nulos en DEPARTAMENTO_RES:', df['DEPARTAMENTO_RES'].isnull().sum())
# cantidad de nulos en PROVINCIA
print('Cantidad de valores nulos en PROVINCIA_PRO:', df['PROVINCIA_PRO'].isnull().sum())
print('Cantidad de valores nulos en PROVINCIA_RES:', df['PROVINCIA_RES'].isnull().sum())
# cantidad de nulos en DISTRITO
print('Cantidad de valores nulos en DISTRITO_PRO:', df['DISTRITO_PRO'].isnull().sum())
print('Cantidad de valores nulos en DISTRITO_RES:', df['DISTRITO_RES'].isnull().sum())
````

````python
# agrupar DOC_PERSONA y DEPARTAMENTO, PROVINCIA Y DISTRITO en una dataframe nuevo
df_procedencia = df[['COD_PERSONA', 'DEPARTAMENTO_PRO', 'PROVINCIA_PRO', 'DISTRITO_PRO']].drop_duplicates()
# contar los nulos de DEPARTAMENTO por COD_PERSONA
nulos_procedencia = df_procedencia[df_procedencia[['DEPARTAMENTO_PRO', 'PROVINCIA_PRO', 'DISTRITO_PRO']].isnull().any(axis=1)]
print('Cantidad de COD_PERSONA con DEPARTAMENTO_PRO nulo:', nulos_procedencia['COD_PERSONA'].nunique())


df_residencia = df[['COD_PERSONA', 'DEPARTAMENTO_RES', 'PROVINCIA_RES', 'DISTRITO_RES']].drop_duplicates()
# contar los nulos de DEPARTAMENTO por COD_PERSONA
nulos_residencia = df_residencia[df_residencia[['DEPARTAMENTO_RES', 'PROVINCIA_RES', 'DISTRITO_RES']].isnull().any(axis=1)]
print('Cantidad de COD_PERSONA con DEPARTAMENTO_RES nulo:', nulos_residencia['COD_PERSONA'].nunique())
````

Dado que **no es posible imputar valores de ubicación** sin contar con otra variable relacionada que proporcione contexto geográfico, la única alternativa viable es **realizar la imputación entre las mismas variables**.  
En este sentido, si para una misma persona (`COD_PERSONA`) una de las variables de ubicación presenta un valor nulo mientras la otra no, se procederá a **rellenar el valor faltante utilizando la información disponible** del mismo individuo.  
De esta manera, se asegura la coherencia interna de los datos sin introducir información externa o artificial. 


````python
# Cuantas personas tienen nulo Departamento en solo PROCEDENCIA , departamento solo RESIDENCIA y ambos
solo_procedencia = nulos_procedencia['COD_PERSONA'].nunique()
solo_residencia = nulos_residencia['COD_PERSONA'].nunique()
ambos = df[df['COD_PERSONA'].isin(nulos_procedencia['COD_PERSONA']) & df['COD_PERSONA'].isin(nulos_residencia['COD_PERSONA'])]['COD_PERSONA'].nunique()

print('Cantidad de COD_PERSONA con solo DEPARTAMENTO_PRO nulo:', solo_procedencia-ambos)
print('Cantidad de COD_PERSONA con solo DEPARTAMENTO_RES nulo:', solo_residencia-ambos)
print('Cantidad de COD_PERSONA con ambos DEPARTAMENTO_PRO y DEPARTAMENTO_RES nulos:' , ambos)

print('Total de personas (COD_PERSONA) a eliminar:', ambos)
print('Porcentaje del total de personas:', (ambos / df['COD_PERSONA'].nunique()) * 100)
````

Con esta información, se observa que solo se perderán datos de **22 individuos**, lo que representa **menos del 1% del total de personas**. Por lo tanto, se considera razonable **aplicar la imputación propuesta** y **eliminar estos 22 individuos con sus registros**, ya que su impacto en el conjunto de datos es mínimo y no compromete la representatividad del análisis.


````python
# 3️⃣ Subconjunto de columnas relevantes sin repeticiones
pro_cols = ['DEPARTAMENTO_PRO', 'PROVINCIA_PRO', 'DISTRITO_PRO']
res_cols = ['DEPARTAMENTO_RES', 'PROVINCIA_RES', 'DISTRITO_RES']

# Crear df_loc sin repeticiones por persona
df_loc = (
    df[['COD_PERSONA'] + pro_cols + res_cols]
    .drop_duplicates(subset='COD_PERSONA', keep='first')
    .copy()
)
df_loc.info()

# 4️⃣ Imputación por persona (bloque completo)
def imputar_bloque(grupo):
    pro_vacios = grupo[['DEPARTAMENTO_PRO', 'PROVINCIA_PRO', 'DISTRITO_PRO']].isna().all().all()
    res_vacios = grupo[['DEPARTAMENTO_RES', 'PROVINCIA_RES', 'DISTRITO_RES']].isna().all().all()

    # Caso 1: No hay procedencia, pero sí residencia -> copiar residencia a procedencia
    if pro_vacios and not res_vacios:
        valores = grupo[['DEPARTAMENTO_RES', 'PROVINCIA_RES', 'DISTRITO_RES']].dropna().iloc[0]
        for col_pro, val in zip(['DEPARTAMENTO_PRO', 'PROVINCIA_PRO', 'DISTRITO_PRO'], valores):
            grupo[col_pro] = val

    # Caso 2: No hay residencia, pero sí procedencia -> copiar procedencia a residencia
    elif res_vacios and not pro_vacios:
        valores = grupo[['DEPARTAMENTO_PRO', 'PROVINCIA_PRO', 'DISTRITO_PRO']].dropna().iloc[0]
        for col_res, val in zip(['DEPARTAMENTO_RES', 'PROVINCIA_RES', 'DISTRITO_RES'], valores):
            grupo[col_res] = val

    return grupo

df_loc_imputado = df_loc.groupby('COD_PERSONA', group_keys=False).apply(imputar_bloque)

df_loc_imputado.info()

# 5️⃣ Detectamos personas que siguen sin procedencia ni residencia
mask_sin_info = (
    df_loc_imputado[['DEPARTAMENTO_PRO', 'PROVINCIA_PRO', 'DISTRITO_PRO']].isna().all(axis=1) &
    df_loc_imputado[['DEPARTAMENTO_RES', 'PROVINCIA_RES', 'DISTRITO_RES']].isna().all(axis=1)
)
# total de eliminados
print('Elementos eliminados:', mask_sin_info.sum())
# total de personas sin info

personas_sin_info = df_loc_imputado.loc[mask_sin_info, 'COD_PERSONA'].unique()

print(f"Personas sin información tras imputar: {len(personas_sin_info)}")

# 6️⃣ Eliminamos esas personas del dataset original
df_final = df.loc[~df['COD_PERSONA'].isin(personas_sin_info)].copy()

print(f"Total de personas eliminadas: {len(personas_sin_info)}")
print(f"Total de registros finales: {df_final.shape[0]}")

# 7️⃣ Fusionamos los datos imputados al dataset completo (con todas las matrículas)
df_final = df_final.drop(columns=['DEPARTAMENTO_PRO', 'PROVINCIA_PRO', 'DISTRITO_PRO',
                                 'DEPARTAMENTO_RES', 'PROVINCIA_RES', 'DISTRITO_RES'], errors='ignore')

df_final = df_final.merge(
    df_loc_imputado,
    on='COD_PERSONA',
    how='left'
)
````

### PARA ESTUDIANTESS

````python
def create_student_features_from_tables(df_estudiante_in, df_matricula, df_curso):
    """
    Calcula las métricas Y de rendimiento (PA, créditos, etc.) usando los
    datos de matrícula y curso, y las añade a la tabla de estudiantes.
    
    Devuelve:
    - df_estudiante_final (con las nuevas columnas Y_ y SEM_CURSADOS)
    - resumen_metricas (solo las métricas, para imputar en test)
    """
    print("Iniciando creación de features de Estudiante (Métricas Y y SEM_CURSADOS)...")
    
    # --- 0. Constante y copia ---
    df_est = df_estudiante_in.copy()
    CREDITOS_REQUERIDOS = 279.0

    # --- 1. Preparar datos de Matrícula ---
    df_m = df_matricula.merge(df_curso[['COD_CURSO', 'CREDITOS']], 
                             on='COD_CURSO', how='left')

    # --- 2. Cálculos de Conteo de Créditos, Cursos y Semestres ---
    df_m['CRED_APRO'] = np.where(df_m['APROBO'] == 'S', df_m['CREDITOS'], 0)
    df_m['CRED_DES'] = np.where(df_m['APROBO'] == 'N', df_m['CREDITOS'], 0)
    df_m['CURSO_APRO'] = np.where(df_m['APROBO'] == 'S', 1, 0)
    df_m['CURSO_DES'] = np.where(df_m['APROBO'] == 'N', 1, 0)
    
    # --- INICIO DE MODIFICACIÓN ---
    # Añadimos el cálculo de 'SEM_CURSADOS' usando nunique()
    stats_df = df_m.groupby('COD_PERSONA').agg(
        CANT_CRED_APRO=('CRED_APRO', 'sum'),
        CANT_CRED_DES=('CRED_DES', 'sum'),
        CANT_CURSOS_APRO=('CURSO_APRO', 'sum'),
        CANT_CURSOS_DES=('CURSO_DES', 'sum'),
        SEM_CURSADOS=('PER_MATRICULA', 'nunique')  # <-- LÍNEA AÑADIDA
    )
    # --- FIN DE MODIFICACIÓN ---

    # --- 3. Cálculo del Promedio Ponderado (PA) ---
    df_m['PUNTOS'] = df_m['NOTA'] * df_m['CREDITOS']
    df_sorted = df_m.sort_values(by='PER_MATRICULA')
    
    df_last_attempt = df_sorted.drop_duplicates(
        subset=['COD_PERSONA', 'COD_CURSO'], 
        keep='last'
    )
    
    pa_grouped = df_last_attempt.groupby('COD_PERSONA')
    suma_puntos = pa_grouped['PUNTOS'].sum()
    suma_creditos = pa_grouped['CREDITOS'].sum()
    
    pa_df = (suma_puntos / suma_creditos).reset_index(name='PA')
    pa_df['PA'] = pa_df['PA'].fillna(0)
    
    # --- 4. Unir todo y calcular porcentajes ---
    resumen_metricas = stats_df.join(pa_df.set_index('COD_PERSONA'))
    
    resumen_metricas['PRCTJ_CRED_APRO'] = \
        (resumen_metricas['CANT_CRED_APRO'] / CREDITOS_REQUERIDOS) * 100
        
    resumen_metricas['PRCTJ_CRED_DES'] = \
        (resumen_metricas['CANT_CRED_DES'] / CREDITOS_REQUERIDOS) * 100

    # --- 5. Fusionar métricas (Y) con la tabla de estudiantes ---
    df_estudiante_final = df_est.merge(
        resumen_metricas.reset_index(),
        on='COD_PERSONA', 
        how='left'
    )
    
    print("Features de Estudiante creadas exitosamente.")
    
    # Rellenar con 0 para alumnos que no tengan historial
    cols_a_rellenar = [col for col in df_estudiante_final.columns if col.startswith('Y_')]
    cols_a_rellenar.append('SEM_CURSADOS')
    df_estudiante_final[cols_a_rellenar] = df_estudiante_final[cols_a_rellenar].fillna(0)
    
    return df_estudiante_final, resumen_metricas.reset_index()
````

````python
def get_stats_block(data):
    """
    Función auxiliar para calcular un bloque de 7 estadísticas
    (AVG, Q25, Q75, MAX, MIN, PRCTJ_S, PRCTJ_N) para un DataFrame dado.
    """
    if data.empty or data['NOTA'].isnull().all():
        return {
            'AVG': np.nan, 'Q25': np.nan, 'Q75': np.nan,
            'MAX': np.nan, 'MIN': np.nan,
            'PRCTJ_S': np.nan, 'PRCTJ_N': np.nan
        }

    avg = data['NOTA'].mean()
    q25 = data['NOTA'].quantile(0.25)
    q75 = data['NOTA'].quantile(0.75)
    max_n = data['NOTA'].max()
    min_n = data['NOTA'].min()

    total_count = len(data)
    aprob_count = (data['APROBO'] == 'S').sum()
    
    prctj_s = (aprob_count / total_count) * 100
    prctj_n = 100.0 - prctj_s

    return {
        'AVG': avg, 'Q25': q25, 'Q75': q75,
        'MAX': max_n, 'MIN': min_n,
        'PRCTJ_S': prctj_s, 'PRCTJ_N': prctj_n
    }

def calculate_course_features(course_data):
    """
    Calcula las 14 características (NG y NPC) para un solo curso.
    Diseñada para ser usada con .apply() en un GroupBy.
    (Esta es tu función, con la corrección de formato)
    """
    
    # 1. Copiar datos (ya no filtra por TIPO_CICLO según tu nuevo código)
    regular_data = course_data.copy()
    
    if regular_data.empty:
        stats_ng = get_stats_block(regular_data)
        stats_npc = get_stats_block(regular_data)
    else:
        # 2. Ordenar por periodo
        regular_data.sort_values(by='PER_MATRICULA', inplace=True)
        unique_periods = regular_data['PER_MATRICULA'].unique()
        
        # 3. Definir NG y NPC
        last_period = unique_periods[-1]
        
        # Corrección para "verano" (ciclos -00)
        if last_period.endswith('-00') and len(unique_periods) > 1:
            last_period = unique_periods[-2] # Usa el penúltimo
            print(f"   (Info: Curso {course_data['CURSO'].iloc[0]} usa {last_period} como NPC)")

        # NPC = Datos del último periodo regular (ya corregido si era verano)
        df_npc = regular_data[regular_data['PER_MATRICULA'] == last_period]
        
        # NG = Todos los periodos EXCEPTO el último (el -00 original si es que hubo)
        ng_periods = unique_periods
        df_ng = regular_data[regular_data['PER_MATRICULA'].isin(ng_periods)]

        # 4. Calcular estadísticas
        stats_ng = get_stats_block(df_ng)
        stats_npc = get_stats_block(df_npc)

    # 5. Crear la serie de resultados
    return pd.Series({
        'AVG_NG': stats_ng['AVG'],
        'quartil-25-NG': stats_ng['Q25'],
        'quartil-75-NG': stats_ng['Q75'],
        'PRCTJE-NG-S': stats_ng['PRCTJ_S'],
        'PRCTJE-NG-N': stats_ng['PRCTJ_N'],
        'max_NG': stats_ng['MAX'],
        'min_NG': stats_ng['MIN'],
        
        'AVG_NPC': stats_npc['AVG'],
        'quartil-25-NPC': stats_npc['Q25'],
        'quartil-75-NPC': stats_npc['Q75'],
        'max_NPC': stats_npc['MAX'],
        'min_NPC': stats_npc['MIN'],
        'PRCTJE-NPC-S': stats_npc['PRCTJ_S'],
        'PRCTJE-NPC-N': stats_npc['PRCTJ_N']
    })


def create_course_features_from_tables(df_curso_in, df_matricula):
    """
    Calcula las métricas NG y NPC usando los datos de matrícula,
    las añade a la tabla de cursos, e imputa los valores faltantes
    (cursos 'cold start') usando el promedio de su FAMILIA.
    
    Argumentos:
    df_curso_in (DataFrame): La malla curricular (71 cursos). 
                             Debe tener 'COD_CURSO' y 'FAMILIA'.
    df_matricula (DataFrame): El historial de matrículas (train set).
    """
    print("Iniciando creación de features de Curso (NG/NPC)...")
    
    # 1. Necesitamos el 'CURSO' (nombre) para el print de depuración
    df_m = df_matricula.merge(df_curso_in[['COD_CURSO', 'CURSO']], 
                             on='COD_CURSO', how='left')
    
    # 2. Calcular estadísticas solo para los cursos con historial
    #    (Esto crea un df con 56 cursos, si usamos tus números de ejemplo)
    course_stats_df = df_m.groupby('COD_CURSO').apply(calculate_course_features)
    
    # 3. Fusionar con la tabla de cursos original (la malla de 71 cursos)
    #    Esto crea NaNs para los 15 cursos sin historial
    df_curso_final = df_curso_in.merge(
        course_stats_df.reset_index(), # reset_index() para usar 'COD_CURSO' como clave
        on='COD_CURSO',
        how='left'
    )
    
    # --- INICIO DE IMPUTACIÓN POR FAMILIA ---
    print("   Imputando features de cursos 'cold start' usando promedios de FAMILIA...")
    
    # 4. Lista de columnas dinámicas que necesitan imputación
    dynamic_cols = [
        'AVG_NG', 'quartil-25-NG', 'quartil-75-NG', 'PRCTJE-NG-S', 
        'PRCTJE-NG-N', 'max_NG', 'min_NG', 
        'AVG_NPC', 'quartil-25-NPC', 'quartil-75-NPC', 'max_NPC', 'min_NPC', 
        'PRCTJE-NPC-S', 'PRCTJE-NPC-N'
    ]
    
    # 5. Calcular las estadísticas promedio POR FAMILIA
    #    Esto usa solo los 56 cursos que SÍ tienen datos
    family_stats = df_curso_final.groupby('FAMILIA')[dynamic_cols].mean()
    
    # 6. Rellenar los NaNs
    for col in dynamic_cols:
        # 6.1. Crea un "mapa" de Familia -> Valor Promedio (ej. 'CS' -> 13.5)
        fill_map = df_curso_final['FAMILIA'].map(family_stats[col])
        
        # 6.2. Rellena los NaNs en la columna usando ese mapa
        df_curso_final[col] = df_curso_final[col].fillna(fill_map)
        
    # 7. Fallback (Respaldo)
    #    Si una FAMILIA entera era nueva (ej. 'ID'), sus promedios en family_stats
    #    serán NaN. Rellenamos esos NaNs finales con el promedio general.
    overall_stats = df_curso_final[dynamic_cols].mean()
    df_curso_final = df_curso_final.fillna(overall_stats)
    
    print("   Imputación completada.")
    # --- FIN DE IMPUTACIÓN POR FAMILIA ---
    
    print("Features de Curso creadas exitosamente.")
    return df_curso_final

````

````python
import pandas as pd
import numpy as np

def verificar_consistencia_funcional(df, key_cols, features_a_verificar):
    """
    Verifica si una clave (key_cols) determina funcionalmente a una lista
    de atributos (features_a_verificar) en un DataFrame.
    
    Devuelve True si es consistente, False si hay errores.
    """
    print(f"--- Verificando Consistencia ---")
    print(f"Clave: {key_cols}")
    
    alumnos_con_problemas = set()
    problemas_encontrados = False

    for feature in features_a_verificar:
        # No tiene sentido verificarse a sí mismo si ya está en la clave
        if feature in key_cols and len(features_a_verificar) > len(key_cols):
             if feature == features_a_verificar[0]: # Solo imprimir una vez
                print(f"   (Omitiendo atributos que son parte de la clave)")
             continue
        
        df_check = df[key_cols + [feature]].drop_duplicates()
        
        inconsistencias = df_check[df_check.duplicated(subset=key_cols, keep=False)]
        
        if inconsistencias.empty:
            print(f"✅ Atributo '{feature}': OK. Es consistente.")
        else:
            problemas_encontrados = True
            alumnos_afectados = 0
            if 'COD_PERSONA' in inconsistencias.columns:
                alumnos_afectados = inconsistencias['COD_PERSONA'].nunique()
                alumnos_con_problemas.update(inconsistencias['COD_PERSONA'].unique())
            
            print(f"❌ Atributo '{feature}': ¡ERROR DE CONSISTENCIA!")
            print(f"   Se encontraron {alumnos_afectados} alumnos con la misma clave pero diferentes '{feature}'.")
            
            print("   --- Ejemplo de inconsistencia ---")
            print(inconsistencias.sort_values(by=key_cols).head(4))
            print("-" * 30)

    print("\n--- Resumen de Verificación ---")
    if not problemas_encontrados:
        print("¡Excelente! Todos los atributos son consistentes con la clave.")
    else:
        print(f"Total de alumnos únicos con alguna inconsistencia: {len(alumnos_con_problemas)}")
    
    # Devolvemos el estado de la consistencia
    return not problemas_encontrados

````

````python
def split_and_prepare_data(df_main, target_cycle, df_curso_original):
    """
    Divide el DataFrame principal en train/test basado en un ciclo,
    valida la consistencia, normaliza en 3 tablas (est, cur, mat),
    crea features de ingeniería solo con datos de train, y prepara
    el conjunto de test para la predicción.
    """
    
    # --- 1. Verificación de Nulos ---
    print(f"--- Iniciando Proceso para Target Cycle: {target_cycle} ---")
    if df_main.isnull().values.any():
        print("¡ERROR! Se encontraron valores nulos en el DataFrame.")
        print("Por favor, limpie los NaNs antes de continuar.")
        return (None,) * 4 # Devuelve 4 Nones
    print("✅ Verificación de Nulos: OK.")

    # calcular un ciclo antes del target_cycle
    year, term = target_cycle.split('-')
    year = int(year)
    term = int(term)
    if term == 0:
        year -= 1
    else:
        term -= 1
    ciclo_anterior = f"{year}-0{term}"
    print(f"Ciclo anterior calculado: {ciclo_anterior}")

    # --- 2. Separación Train/Test/Ignorados ---
    df_main = df_main[df_main['PER_INGRESO'] < ciclo_anterior].copy() # Filtrar solo alumnos ingresados antes del target_cycle
    df_train = df_main[df_main['PER_MATRICULA'] <= target_cycle].copy()
    df_test = df_main[df_main['PER_MATRICULA'] == target_cycle].copy()
    
    ignored_count = len(df_main[df_main['PER_MATRICULA'] > target_cycle])
    print(f"Separación de datos: {len(df_train)} filas de train, {len(df_test)} filas de test.")
    if ignored_count > 0:
        print(f"   (Se ignoraron {ignored_count} filas posteriores a {target_cycle})")
        
    if df_train.empty:
        print(f"¡ERROR! No hay datos de entrenamiento (anteriores a {target_cycle}).")
        return (None,) * 4
    if df_test.empty:
        print(f"¡ADVERTENCIA! No hay datos de prueba (iguales a {target_cycle}).")
        # Continuamos, pero el test set estará vacío.

    # --- 3. Validación de Consistencia (Solo en Train) ---
    print("\n--- Iniciando Validación de Consistencia en Train ---")
    
    # Claves y features para Estudiante
    clave_estudiante = ['COD_PERSONA']
    features_estudiante = [
        'COD_PERSONA', 'SEXO', 'PER_INGRESO', 'ESTADO_CIVIL', 'TIPO_COLEGIO',
        'PTJE_INGRESO', 
        'CANT_RESERVAS', 'BECA_VIGENTE',
        'FECHA_NACIMIENTO', 'POBREZA_RES', 'POBREZA_PRO'
    ]
    ok_est = verificar_consistencia_funcional(df_train, clave_estudiante, features_estudiante)

    # Claves y features para Curso
    clave_curso = ['COD_CURSO']
    features_curso = [
        'COD_CURSO', 'CURSO', 'CREDITOS', 'HRS_CURSO', 'TIPO_CURSO', 'FAMILIA', 'NIVEL_CURSO'
    ]
    ok_cur = verificar_consistencia_funcional(df_train, clave_curso, features_curso)

    # Claves y features para Matrícula
    clave_matricula = ['COD_PERSONA', 'COD_CURSO', 'PER_MATRICULA']
    features_matricula = [
        'NOTA', 'HRS_INASISTENCIA', 'PRCTJE_INASISTENCIA', 'APROBO', 'ESTADO', 'PONDERADO' , 'TIPO_CICLO'
    ]
    # Agregamos las claves a la lista de features para la función
    features_matricula_check = clave_matricula + features_matricula
    ok_mat = verificar_consistencia_funcional(df_train, clave_matricula, features_matricula_check)

    if not (ok_est and ok_cur and ok_mat):
        print("\n¡ERROR! Falló la verificación de consistencia funcional.")
        print("Por favor, revise los errores y limpie los datos de train.")
        return (None,) * 4
    
    print("\n✅ Verificación de Consistencia en Train: OK.")

    # --- 4. Normalización de Tablas (Train) ---
    print("Normalizando datos de train en 3 tablas...")
    
    df_estudiante_train = df_train[features_estudiante].drop_duplicates(subset=clave_estudiante)
    # df_curso_train = df_train[features_curso].drop_duplicates(subset=clave_curso)
    df_curso_train = df_curso_original.copy()  # Usar la malla curricular original para cursos

    # La tabla de matrícula necesita las claves y las features
    df_matricula_train = df_train[clave_matricula + features_matricula].drop_duplicates()
    
    print(f"   Tabla Estudiante (train): {df_estudiante_train.shape}")
    print(f"   Tabla Curso (train):      {df_curso_train.shape}")
    print(f"   Tabla Matrícula (train):  {df_matricula_train.shape}")

    # --- 5. Feature Engineering (Usando solo Train) ---
    
    # 5.1. Features de Estudiante (Y_ metrics)
    df_estudiante_train, student_metrics = create_student_features_from_tables(
        df_estudiante_train, df_matricula_train, df_curso_train
    )
    
    # 5.2. Features de Curso (NG/NPC metrics)
    df_curso_train = create_course_features_from_tables(
        df_curso_train, df_matricula_train
    )

    # --- 6. Preparación del Test Set ---
    print("\nPreparando el conjunto de Test...")
    
    # 6.1. Seleccionar solo las columnas de matrícula
    df_test_prepared = df_test[clave_matricula + features_matricula].copy()
    
    # # 6.2. Eliminar la variable objetivo
    # if 'NOTA' in df_test_prepared.columns:
    #     df_test_prepared = df_test_prepared.drop('NOTA', axis=1)
    #     print("   'NOTA' eliminada del test set.")
    
    # 6.3. Imputar valores basados en el historial (train)
    
    # Calcular promedios históricos del train
    hist_inasistencia = df_train.groupby('COD_PERSONA')['HRS_INASISTENCIA'].mean()
    hist_prctje_inasis = df_train.groupby('COD_PERSONA')['PRCTJE_INASISTENCIA'].mean()
    
    # Usar el PA calculado (Y_PA) desde el resumen de métricas
    pa_historico = student_metrics.set_index('COD_PERSONA')['PA']

    print("   Imputando 'HRS_INASISTENCIA', 'PRCTJE_INASISTENCIA' y 'PONDERADO' en test...")
    
    # Mapear valores
    df_test_prepared['HRS_INASISTENCIA'] = df_test_prepared['COD_PERSONA'].map(hist_inasistencia)
    df_test_prepared['PRCTJE_INASISTENCIA'] = df_test_prepared['COD_PERSONA'].map(hist_prctje_inasis)
    df_test_prepared['PONDERADO'] = df_test_prepared['COD_PERSONA'].map(pa_historico)
    
    # Rellenar con 0 para alumnos nuevos (sin historial en train)
    nulos_antes = df_test_prepared[['HRS_INASISTENCIA', 'PONDERADO']].isnull().sum().sum()
    if nulos_antes > 0:
        print(f"   (Se encontraron {nulos_antes} valores nulos por alumnos nuevos. Rellenando con 0.)")
        df_test_prepared['HRS_INASISTENCIA'] = df_test_prepared['HRS_INASISTENCIA'].fillna(0)
        df_test_prepared['PRCTJE_INASISTENCIA'] = df_test_prepared['PRCTJE_INASISTENCIA'].fillna(0)
        df_test_prepared['PONDERADO'] = df_test_prepared['PONDERADO'].fillna(0)

    print(' Imputando ESTADO en test...')
    df_test_prepared['ESTADO'] = df_test_prepared['COD_PERSONA'].map(df_train.groupby('COD_PERSONA')['ESTADO'].first())

    # ========== Generar edad en df_estudiante_train
    # 1. calcular año actual del target_cycle
    anio_actual = int(target_cycle.split('-')[0])
    # 2. calcular año de nacimiento
    df_estudiante_train['ANIO_NACIMIENTO'] = pd.to_datetime(df_estudiante_train['FECHA_NACIMIENTO']).dt.year
    # 3. calcular edad  
    df_estudiante_train['EDAD'] = anio_actual - df_estudiante_train['ANIO_NACIMIENTO']
    # 4. eliminar ANIO_NACIMIENTO
    df_estudiante_train = df_estudiante_train.drop('ANIO_NACIMIENTO', axis=1)
    print("   'EDAD' calculada y añadida a la tabla de Estudiante.")

    # ========== generar TIPO_CICLO con target_cycle
    df_test_prepared['TIPO_CICLO'] = np.where(df_test_prepared['PER_MATRICULA'].str.endswith('-00'), 'VERANO', 'REGULAR')
    print("   'TIPO_CICLO' generada en el test set.")

    # --- INICIO DE LA CORRECCIÓN DE 'ESTADO' ---
    # 6.4. Generar 'ESTADO_ANTERIOR' como feature
    print("   Imputando 'ESTADO_ANTERIOR' en test...")

    # 1. Ordenar el train set cronológicamente
    df_train_sorted = df_train.sort_values(by='PER_MATRICULA')

    # 2. Obtener el ÚLTIMO estado conocido de cada alumno del historial
    #    Esto nos da el estado más reciente ANTES del target_cycle
    last_known_estado = df_train_sorted.groupby('COD_PERSONA')['ESTADO'].last()

    # 3. Mapear este estado al test set
    df_test_prepared['ESTADO'] = df_test_prepared['COD_PERSONA'].map(last_known_estado)


    print("   'ESTADO_ANTERIOR' generado exitosamente.")
    # --- FIN DE LA CORRECCIÓN DE 'ESTADO' ---
        # 

    print("\n--- ¡Proceso Completado! ---")

    # ELIMINAR APROBO DEL TEST
    if 'APROBO' in df_test_prepared.columns:
        df_test_prepared = df_test_prepared.drop('APROBO', axis=1)
        print("   'APROBO' eliminada del test set.")
    
    #   eliminar aprobo de la tabla de matricula de train tambien
    if 'APROBO' in df_matricula_train.columns:
        df_matricula_train = df_matricula_train.drop('APROBO', axis=1)
        print("   'APROBO' eliminada del train set de matricula.")
    
    # --- 7. Devolver las 4 tablas ---
    return df_estudiante_train, df_curso_train, df_matricula_train, df_test_prepared
````

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# --- 1. Definir las Columnas para el Modelo (Forma Automática) ---
# Asumiendo que X_train, y_train, X_test, y_test ya existen en tu entorno

# Columnas que NO deben ser features (ej: IDs, fechas ya procesadas, etc.)
# Estas columnas serán ignoradas por el modelo.
excluded_features = [
    'COD_PERSONA',      # Es un identificador, no un feature predictivo
    'PER_MATRICULA',    # FUGA DE DATOS (Time Leakage)
    'FECHA_NACIMIENTO'  # Ya fue convertida a 'EDAD'
]

# Identificar automáticamente las columnas por su tipo de dato (dtype)
# Esto es mucho más seguro que escribirlas a mano.

# Seleccionamos todas las columnas numéricas
numerical_features = X_train.select_dtypes(include=np.number).columns.to_list()

# Seleccionamos todas las columnas categóricas (object o category)
categorical_features = X_train.select_dtypes(include=['object', 'category']).columns.to_list()

# Ahora, eliminamos las columnas que no queremos de nuestras listas de features
numerical_features = [col for col in numerical_features if col not in excluded_features]
categorical_features = [col for col in categorical_features if col not in excluded_features]

# (Opcional) Imprime las listas para confirmar que son correctas
print("--- Features Numéricos Detectados ---")
print(numerical_features)
print("\n--- Features Categóricos Detectados ---")
print(categorical_features)
print("-" * 30)


# --- 2. Crear el Pipeline de Preprocesamiento ---
# (Esta sección es idéntica a la anterior, pero ahora usa las listas autogeneradas)

# Pipeline para features numéricos
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Pipeline para features categóricos
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Combinar ambos pipelines usando ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='drop'  # Ignora cualquier columna que no esté en las listas (como las excluidas)
)

# --- 3. Definir los Modelos ---
# (Idéntico a antes)
models = {
    'Linear Regression': LinearRegression(),
    
    # Reducimos la profundidad para combatir el overfitting
    'Random Forest': RandomForestRegressor(n_estimators=100, 
                                           random_state=42, 
                                           n_jobs=-1, 
                                           max_depth=7), # <-- CAMBIO DE 10 a 7
    
    'LightGBM': LGBMRegressor(random_state=42, 
                              n_jobs=-1,
                              max_depth=7) # <-- AÑADIDO para controlar
}

# --- 4. Entrenar y Evaluar los Modelos ---
# (Idéntico a antes)

results = {}

for model_name, model in models.items():
    
    print(f"--- Entrenando {model_name} ---")
    
    # Crear el pipeline completo: Preprocesador + Modelo
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])
    
    # Entrenar el modelo
    pipeline.fit(X_train, y_train)
    
    # Predecir en datos de entrenamiento y prueba
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)
    
    # Calcular métricas
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    train_mae = mean_absolute_error(y_train, y_train_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    # Guardar resultados
    results[model_name] = {
        'Train R2': train_r2,
        'Test R2': test_r2,
        'Train MAE': train_mae,
        'Test MAE': test_mae,
        'Train RMSE': train_rmse,
        'Test RMSE': test_rmse
    }
    
    print(f"Resultados para {model_name}:")
    print(f"  Train R2: {train_r2:.4f} | Test R2: {test_r2:.4f}")
    print(f"  Train MAE: {train_mae:.4f} | Test MAE: {test_mae:.4f}")
    print(f"  Train RMSE: {train_rmse:.4f} | Test RMSE: {test_rmse:.4f}")
    print("-" * 30)

# --- 5. Mostrar Resumen de Resultados ---
# (Idéntico a antes)

print("\n--- Resumen de Métricas (Test) ---")
results_df = pd.DataFrame(results).T
print(results_df[['Test R2', 'Test MAE', 'Test RMSE']].sort_values(by='Test R2', ascending=False))
````
